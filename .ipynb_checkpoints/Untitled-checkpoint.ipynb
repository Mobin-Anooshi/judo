```python
def extract_match_info(contest):
    players = []
    for player_div in contest.find_elements(By.CSS_SELECTOR, "div.flex.items-center.gap-2.flex-1"):
        try:
            country = player_div.find_element(By.CSS_SELECTOR, "span.text-xs").text
            name_element = player_div.find_element(By.CSS_SELECTOR, "a")
            name = name_element.text.replace("\n", " ").strip()
            players.append({"name": name, "country": country})
        except:
            players.append(None)  # در صورت خالی بودن (Bye)
    
    # شناسایی برنده
    winner = None
    if len(players) == 2 and players[0] and players[1]:
        try:
            white_line = contest.find_element(By.CSS_SELECTOR, "div.bg-white")
            winner = players[0]["name"] if white_line else players[1]["name"]
        except:
            winner = None
    elif players[0]:
        winner = players[0]["name"]
    elif players[1]:
        winner = players[1]["name"]

    return {
        "player1": players[0],
        "player2": players[1] if len(players) > 1 else None,
        "winner": winner
    }

# استخراج داده‌ها
tournament_results = {"groups": {}, "champion": None}
group_titles = driver.find_elements(By.CSS_SELECTOR, "div.bracket-title")
bracket_columns = driver.find_elements(By.CSS_SELECTOR, "div.bracket-column")

for idx, (title_el, column) in enumerate(zip(group_titles, bracket_columns)):
    group_title = title_el.text.strip()
    matches = []
    contests = column.find_elements(By.CSS_SELECTOR, "div.contest")

    for contest in contests:
        match_info = extract_match_info(contest)
        matches.append(match_info)

    if group_title not in tournament_results["groups"]:
        tournament_results["groups"][group_title] = []

    tournament_results["groups"][group_title].extend(matches)

# شناسایی قهرمان
for group_name in tournament_results["groups"]:
    for match in tournament_results["groups"][group_name]:
        if match["player1"] and match["player2"] and match["player2"]["name"] == "Bye":
            # احتمال زیاد فینال بوده
            tournament_results["champion"] = {
                "name": match["winner"],
                "country": match["player1"]["country"] if match["player1"]["name"] == match["winner"] else match["player2"]["country"]
            }

# ذخیره فایل
with open("tournament_grouped_results.json", "w", encoding="utf-8") as f:
    json.dump(tournament_results, f, indent=2, ensure_ascii=False)

# چاپ خروجی
for group, matches in tournament_results["groups"].items():
    print(f"\nGroup: {group}")
    for match in matches:
        player1 = match["player1"]["name"] if match["player1"] else "Bye"
        player2 = match["player2"]["name"] if match["player2"] else "Bye"
        winner = match["winner"] if match["winner"] else "Unknown"
        print(f"Match: {player1} vs {player2} -> Winner: {winner}")

if tournament_results["champion"]:
    champ = tournament_results["champion"]
    print(f"\nChampion: {champ['name']} ({champ['country']})")
```

```python
def extract_match_info(contest):
    players = []
    for player_div in contest.find_elements(By.CSS_SELECTOR, "div.flex.items-center.gap-2.flex-1"):
        try:
            country = player_div.find_element(By.CSS_SELECTOR, "span.text-xs").text
            name_element = player_div.find_element(By.CSS_SELECTOR, "a")
            name = name_element.text.replace("\n", " ").strip()
            players.append({"name": name, "country": country})
        except:
            players.append(None)  # در صورت خالی بودن (Bye)
    
    # شناسایی برنده (بر اساس حضور در مرحله بعدی یا خط سفید/آبی)
    winner = None
    if len(players) == 2 and players[0] and players[1]:
        try:
            white_line = contest.find_element(By.CSS_SELECTOR, "div.bg-white")
            winner = players[0]["name"] if white_line else players[1]["name"]
        except:
            winner = None  # اگر نتوان برنده را تشخیص داد
    elif players[0]:
        winner = players[0]["name"]  # Bye
    elif players[1]:
        winner = players[1]["name"]  # Bye

    return {
        "player1": players[0],
        "player2": players[1] if len(players) > 1 else None,
        "winner": winner
    }

# استخراج تمام مراحل
tournament_results = {"rounds": [], "champion": None}

# فرض: براکت مربوط به Pool A و -50 kg
pool_name = "Pool A"
weight_class = "-50 kg"

# پیدا کردن براکت‌های مربوط به Pool A و -50 kg
try:
    # فرض: هر Pool یا دسته وزنی در یک المان با کلاس خاص (مثل pool-title یا weight-class) مشخص شده
    pool_element = driver.find_element(By.XPATH, f"//div[contains(text(), '{pool_name}')]")
    weight_element = driver.find_element(By.XPATH, f"//div[contains(text(), '{weight_class}')]")
    bracket_container = pool_element.find_element(By.XPATH, "./ancestor::div[contains(@class, 'bracket')]")
except:
    # اگر Pool یا وزن پیدا نشد، کل صفحه را به‌عنوان براکت در نظر بگیر
    bracket_container = driver

bracket_columns = bracket_container.find_elements(By.CSS_SELECTOR, "div.bracket-column")

for col_idx, column in enumerate(bracket_columns[:-1]):  # نادیده گرفتن ستون آخر (قهرمان)
    # تعیین نام مرحله بر اساس تعداد مسابقات
    contests = column.find_elements(By.CSS_SELECTOR, "div.contest")
    num_matches = len(contests)
    
    if num_matches >= 4:
        round_name = f"{pool_name} {weight_class} - First Round"
    elif num_matches == 2:
        round_name = f"{pool_name} {weight_class} - Semi Finals"
    elif num_matches == 1:
        round_name = f"{pool_name} {weight_class} - Final"
    else:
        round_name = f"{pool_name} {weight_class} - Unknown Round"

    matches = []
    for contest in contests:
        match_info = extract_match_info(contest)
        matches.append(match_info)
    
    tournament_results["rounds"].append({"name": round_name, "matches": matches})

# شناسایی قهرمان (آخرین مسابقه در ستون آخر)
final_round = tournament_results["rounds"][-1]
if final_round["matches"]:
    champion_name = final_round["matches"][0]["winner"]
    champion_country = final_round["matches"][0]["player1"]["country"] if final_round["matches"][0]["player1"] and final_round["matches"][0]["player1"]["name"] == champion_name else final_round["matches"][0]["player2"]["country"]
    tournament_results["champion"] = {"name": champion_name, "country": champion_country}

# ذخیره خروجی به‌صورت JSON
with open("tournament_results.json", "w", encoding="utf-8") as f:
    json.dump(tournament_results, f, indent=2, ensure_ascii=False)

# چاپ نتایج برای بررسی
for round_data in tournament_results["rounds"]:
    print(f"\nStage: {round_data['name']}")
    for match in round_data["matches"]:
        player1 = match["player1"]["name"] if match["player1"] else "Bye"
        player2 = match["player2"]["name"] if match["player2"] else "Bye"
        winner = match["winner"] if match["winner"] else "Unknown"
        print(f"Match: {player1} vs {player2} -> Winner: {winner}")

print(f"\nChampion: {tournament_results['champion']['name']} ({tournament_results['champion']['country']})")
```

```python
import json
from collections import defaultdict
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException

def process_raw_data(raw_data):
    lines = [line.strip() for line in raw_data.splitlines() if line.strip()]
    current_stage = None
    current_round = None
    matches = []
    last_round_matches = defaultdict(list)

    stage_name_mapping = {
        'POOL A': 'Pool A',
        'POOL B': 'Pool B',
        'POOL C': 'Pool C',
        'POOL D': 'Pool D',
        'REPECHAGE': 'Repechage',
        'SEMI-FINALS': 'Semi-Finals',
        'FINALS': 'Finals'
    }

    round_title_mapping = {
        '1-16': 'Round of 32',
        '1-8': 'Round of 16',
        '1-4': 'Quarter-Finals',
        '1-2': 'Semi-Finals'
    }

    for line in lines:
        line = line.strip()
        if line == '----':
            continue
        if line in stage_name_mapping:
            current_stage = stage_name_mapping[line]
        elif line.startswith('round'):
            current_round = line.split()[1]
        elif line.startswith('[') and current_stage and current_round:
            try:
                match_players = eval(line)
                match_info = {
                    "round": current_round,
                    "players": match_players,
                    "winner": None
                }
                matches.append((current_stage, match_info))
                last_round_matches[current_stage, current_round].append(match_players)
            except:
                print(f"Error parsing players: {line}")

    def find_winner(current_players, next_round_players):
        if len(current_players) == 1:
            return current_players[0]
        current_names = {p['name'] for p in current_players}
        next_names = {player['name'] for match in next_round_players for player in match}
        for player in current_players:
            if player['name'] in next_names:
                return player
        return None

    staged_matches = defaultdict(list)
    stage_rounds = defaultdict(list)
    for stage, match in matches:
        stage_rounds[stage].append(match['round'])

    for stage in stage_rounds:
        stage_rounds[stage] = sorted(list(set(stage_rounds[stage])), key=lambda x: int(x.split('-')[-1]), reverse=True)

    for stage, match in matches:
        round_ = match['round']
        current_players = match['players']
        rounds_in_stage = stage_rounds[stage]
        is_last_round = rounds_in_stage.index(round_) == 0
        match['round_title'] = 'Final' if is_last_round and len(current_players) == 1 else round_title_mapping.get(round_, round_)

        current_round_idx = rounds_in_stage.index(round_)
        next_round = rounds_in_stage[current_round_idx - 1] if current_round_idx > 0 else None
        next_round_players = last_round_matches.get((stage, next_round), []) if next_round else []
        match['winner'] = find_winner(current_players, next_round_players)

        staged_matches[stage].append(match)

    return dict(staged_matches)

def process_data(data_list):
    result = []
    for data in data_list:
        items = [item.strip() for item in data.split('\n') if item.strip()]
        for i in range(0, len(items), 2):
            if i + 1 < len(items):
                result.append({
                    'country': items[i],
                    'name': items[i + 1]
                })
    return result

def generate_raw_data(driver):
    raw_data_lines = []
    try:
        stages_container = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//div[contains(@class, "tournament-bracket")]'))
        )
        child_divs_stages = stages_container.find_elements(By.XPATH, './div')
        print(f"Found {len(child_divs_stages)} stages")

        match_count_to_round = {
            16: '1-32',
            8: '1-16',
            4: '1-8',
            2: '1-4',
            1: '1-2'
        }

        for stage_idx, stage_div in enumerate(child_divs_stages, 1):
            try:
                stage_name_elem = WebDriverWait(driver, 5).until(
                    EC.presence_of_element_located((
                        By.XPATH, 
                        f'//div[contains(@class, "tournament-bracket")]//div[{stage_idx}]//h3'
                    ))
                )
                stage_name = stage_name_elem.text.strip().upper()
                if stage_name not in stage_name_mapping:
                    print(f"Unknown stage name: {stage_name}, skipping")
                    continue
                raw_data_lines.append(stage_name)
                print(f"Processing stage: {stage_name}")

                rounds_container = stage_div.find_element(By.XPATH, './/div[contains(@class, "rounds")]')
                round_divs = rounds_container.find_elements(By.XPATH, './div')

                for round_div in round_divs:
                    matches = round_div.find_elements(By.XPATH, './div')
                    rr = len(matches)
                    round_name = match_count_to_round.get(rr, f'1-{rr*2}')
                    raw_data_lines.append(f'round {round_name}')
                    print(f"Round {round_name} with {rr} matches")

                    for match in matches:
                        aa = [match.text]
                        players = process_data(aa)
                        if players:
                            raw_data_lines.append(json.dumps(players, ensure_ascii=False))
                        else:
                            print(f"No players found for match in stage {stage_name}, round {round_name}")

                raw_data_lines.append('----')

            except (NoSuchElementException, TimeoutException) as e:
                print(f"Error processing stage {stage_idx}: {e}")
                continue

    except TimeoutException:
        print("Failed to load stages container")
        return ""

    return '\n'.join(raw_data_lines)

if __name__ == "__main__":
    # فرض می‌کنیم driver تعریف شده
    # from selenium import webdriver
    # driver = webdriver.Chrome()
    # driver.get("your_website_url")

    try:
        raw_data = generate_raw_data(driver)
        print("Raw Data Generated:")
        print(raw_data)

        result = process_raw_data(raw_data)

        with open('matches_by_stage.json', 'w', encoding='utf-8') as f:
            json.dump(result, f, indent=2, ensure_ascii=False)

        print("✅ فایل matches_by_stage.json ساخته شد!")
    finally:
        # driver.quit()
        pass
```

```python
# print(player_id)

for p_id in range(1) :
    # driver.get(f'https://judotv.com/judoka/{p_id}')
    driver.get(f'https://judotv.com/judoka/79453')
    sleep(2)
    player_name = driver.find_element(By.XPATH ,'/html/body/div[1]/div/div[2]/div/main/div[2]/div[1]/div/div/div[1]/div[2]/div[2]/h1').text
    try:
        while True :
            judo_button = driver.find_element(By.XPAplayer_idTH,'/html/body/div[1]/div/div[2]/div/main/div[2]/div[1]/div/div[5]/button')
            judo_button.click()
            sleep(2)
    except :
        pass
    links = driver.find_elements(By.CSS_SELECTOR, "a[href*='/competitions/'][href*='/contests/']")
    hrefs = [link.get_attribute("href") for link in links]
    print
    all_matches=[]
    for ur in hrefs :
        driver.get(ur)
        
        judoka_links = driver.find_elements(By.CSS_SELECTOR, "a[href*='/judoka/']")
        
        names = []
        seen_ids = set()
        
        for link in judoka_links:
            href = link.get_attribute("href")
            judoka_id = href.split("/")[-1]
        
            if judoka_id not in seen_ids:
                name = link.text.strip().replace("\n", " ")
                if name:
                    names.append(name)
                    seen_ids.add(judoka_id)
        
            if len(names) == 2:
                break 
        # print("بازیکن اول:", names[0])
        # print("بازیکن دوم:", names[1])
        grid = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[2]/div/main/div[2]/div/div[2]/div[2]/div[2]')
        columns = grid.find_elements(By.XPATH, "./div")
        events = []
        i = 0
        while i < len(columns):
            col1 = columns[i].text.strip()
            col2 = columns[i+1].text.strip() if i+1 < len(columns) else ''
            col3 = columns[i+2].text.strip() if i+2 < len(columns) else ''
        
            # تشخیص اینکه متن در کدوم ستون قرار گرفته
            left = col1 if col1 else None
            time = col2 if col2 else None
            right = col3 if col3 else None
        
            # بررسی حالت خاص که ستون وسط دکمه‌ست (یعنی امتیاز بازیکن اول بوده)
            if not time or ":" not in time:
                # پس ستون اول: زمان — ستون دوم: اتفاق سمت راست
                time = col1 if ":" in col1 else None
                left = None
                right = col2
        
            events.append({
                "time": time,
                names[0]: left,
                names[1]: right
            })

            i += 3

            match_data = {
                "judoka_1": names[0],
                "judoka_2": names[1],
                "winner": names[1],  # چون گفتی همیشه دومی برنده‌ست
                "events": events
            }
        all_matches.append(match_data)
    print(json.dumps(all_matches, indent=2, ensure_ascii=False))
    with open(f'{player_name}.json', "w", encoding="utf-8") as f:
        json.dump(all_matches, f, ensure_ascii=False, indent=2)
        
```
